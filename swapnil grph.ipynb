```python
import matplotlib.pyplot as pl

test = [1, 2, 3, 4, 5]
marks = [25, 34, 49, 40, 48]

pl.title('Analysis of Test Marks')
pl.xlabel('Test-No')
pl.ylabel('Marks')
pl.plot(test, marks, 'g', marker='D', markersize=5, markeredgecolor='purple', linestyle='solid')
pl.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\lab-03-19\\Desktop\\Rainfall1.csv")
x=df['Months']
y=df['Rainfall in cm']
plt.figure(figsize=(5,3))
plt.figure(figsize=(6,4))
color= np.array([0,10,20,30,40,45,50,60,70,80,90,100])
plt.scatter(x,y,c=color,cmap='viridis')
plt.title('Rainfall data of odisha',fontname='Calibri',color='m')
plt.xtick(rotation=45)
plt.xlabel('Months',fontname='Calibri',color='b',fontsize=12)
plt.ylabel ('Rainfall in cm',fontname='calibri',color='b',fontsize=12) 
plt.show()
```
